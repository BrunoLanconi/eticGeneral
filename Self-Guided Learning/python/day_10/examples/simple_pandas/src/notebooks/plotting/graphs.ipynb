```python
# Basic plotting
from scipy.stats import norm # type: ignore
import matplotlib.pyplot as plt # type: ignore
import numpy as np # type: ignore

x = np.arange(-3, 3, 0.001)

plt.plot(x, norm.pdf(x))
plt.show()
```

```python
# Plotting multiple plots
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.show()
```

```python
# Save the plot
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.savefig("output/sample.png", format="png")
```

```python
# Adjust the axes, grid and customize the plot
axes = plt.axes()

axes.set_xlim(left=-5, right=5) # X-axis min and max
axes.set_ylim(bottom=0.0, top=1.0) # Y-axis min and max
axes.set_xticks(np.arange(-5, 6, 2)) # The labels on the X-axis
axes.set_yticks(np.arange(0, 1.1, 0.1)) # The labels on the Y-axis

axes.grid() # Show the grid
axes.grid(linestyle='dashed', linewidth=0.5) # Custom grid, you may also use 'solid', 'dotted', 'dashdot' or '-', '--', '-.', ':'

# Labeling the axes and adding a legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.plot(x, norm.pdf(x), 'b-') # 'b-' stands for blue solid line
plt.plot(x, norm.pdf(x, 1.0, 0.5), 'r:') # 'r:' stands for red dotted line

# NOTE: The order of the labels in the legend should match the order of the plots
plt.legend(['Solid blue', 'Dotted red'], loc=4) # loc=4 is the lower right corner, you may also use 0-10
plt.show()
```

```python
# Pie chart
values = [12, 55, 4, 32, 14]
colors = ['r', 'g', 'b', 'c', 'm']
explode = [0, 0, 0.2, 0, 0] # Explode the 3rd slice by 20%
labels = ['India', 'United States', 'Russia', 'China', 'Europe']

plt.pie(values, colors=colors, labels=labels, explode=explode)
plt.title('Student Locations')
plt.show()
```

```python
# Bar chart
plt.bar(range(0, 5), values, color=colors)
plt.title('Student Locations')

# Add ticks to the bars
plt.xticks(range(0, 5), labels)

# Labeling the axes
plt.xlabel('Countries')
plt.ylabel('Number of Students')

plt.show()
```

```python
# Scatter plot
x = np.random.randn(1000) # Generate an array of 1000 random numbers
y = np.random.randn(1000)

plt.scatter(x, y)
plt.show()
```

```python
# Histogram
incomes = np.random.normal(27000, 15000, 10000) # Mean, standard deviation, number of samples

plt.hist(incomes, 50, ) # 50 bins
plt.show()
```

```python
# Box and whisker plot
# The box represents the interquartile range, the line in the middle is the median and the whiskers are the min and max values
uniformSkewed = np.random.rand(100) * 100 - 40
high_outliers = np.random.rand(10) * 50 + 100
low_outliers = np.random.rand(10) * -50 - 100
data = np.concatenate((uniformSkewed, high_outliers, low_outliers))

plt.boxplot(data)
plt.show()
```
