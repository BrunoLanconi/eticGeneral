```python
import pandas as pd

df = pd.read_csv("../../../inputs/employees.csv")
ages = df["Age"]
sample = list(ages)

sample
```

```python
# The mean is the sum of the sample divided by the number of elements in the sample
mean = sum(sample)/len(sample)

mean
```

```python
# The median is the middle value of the sample. If the sample has an even number of elements, the median is the average of the two middle elements.
sample.sort()

median = sample[len(sample)//2]

median
```

```python
# The mode is the value that appears most frequently in the sample.
mode = max(set(sample), key = sample.count)

mode
```

```python
# The population variance is the average of the squared differences from the Mean
population_variance = sum((x - mean) ** 2 for x in sample) / len(sample)

population_variance
```

```python
# If the sample is a subset of the population, the sample variance is calculated by dividing the sum of the squared differences from the Mean by the number of elements in the sample minus 1.
sample_variance = sum((x - mean) ** 2 for x in sample) / (len(sample) - 1)

sample_variance
```

```python
# The standard deviation is the square root of the variance
standard_deviation = population_variance ** 0.5

standard_deviation
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

x = np.linspace(mean - 3*standard_deviation, mean + 3*standard_deviation, 100)
# Probability Density Function (pdf)
pdf = stats.norm.pdf(x, mean, standard_deviation)

plt.plot(x, pdf)
plt.axvline(mean, color='r', linestyle='dashed', linewidth=2)
plt.axvline(mean + 2*standard_deviation, color='g', linestyle='dashed', linewidth=2)
plt.axvline(mean + standard_deviation, color='g', linestyle='dashed', linewidth=2)
plt.axvline(mean - 2*standard_deviation, color='g', linestyle='dashed', linewidth=2)
plt.axvline(mean - standard_deviation, color='g', linestyle='dashed', linewidth=2)

# # Add value labels above the green lines
minus_one_sigma_text = str(round(mean - standard_deviation, 2))
minus_two_sigma_text = str(round(mean - 2*standard_deviation, 2))
plus_one_sigma_text = str(round(mean + standard_deviation, 2))
plus_two_sigma_text = str(round(mean + 2*standard_deviation, 2))

plt.text(mean, pdf.max() * 1.1, str(round(mean, 2)), verticalalignment='center')
plt.text(mean + 2*standard_deviation, pdf.max() * 1.1, plus_two_sigma_text, verticalalignment='center')
plt.text(mean + standard_deviation, pdf.max() * 1.1, plus_one_sigma_text, verticalalignment='center')
plt.text(mean - 2*standard_deviation, pdf.max() * 1.1, minus_two_sigma_text, verticalalignment='center')
plt.text(mean - standard_deviation, pdf.max() * 1.1, minus_one_sigma_text, verticalalignment='center')

plt.show()
```
