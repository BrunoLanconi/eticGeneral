```python
import numpy as np

values = np.random.normal(0, 0.5, 10000)

values
```

```python
# First moment - mean
mean = np.mean(values)

mean
```

```python
# Second moment - variance
variance = np.var(values)

variance
```

```python
# Third moment - skewness - how much the distribution is skewed to the left or right
import scipy.stats as sp

skew = sp.skew(values)

skew
```

```python
# Fourth moment - kurtosis - how much the distribution is peaked or flat
kurtosis = sp.kurtosis(values)

kurtosis
```
